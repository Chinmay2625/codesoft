```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Titanic dataset
data = pd.read_csv('Titanic-Dataset.csv')
```

```python
# Data preprocessing
# Drop rows with missing values
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})
data['Embarked'] = data['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
```

```python
# Define bins for age groups
bins = [0, 12, 20, 40, 60, 100]
labels = ['0-12', '13-20', '21-40', '41-60', '61+']
data['AgeGroup'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)
```

```python
data.head()
```

```python
# Create a DataFrame for correlation heatmap
data_for_heatmap = data[['Pclass', 'Age', 'Embarked', 'Survived']].copy()
```

```python
# Plotting
plt.figure(figsize=(12, 18))

# Heatmap
plt.subplot(4, 1, 1)
sns.heatmap(data_for_heatmap.corr(), annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Correlation Heatmap')
```

```python
# Bar plot for Survival vs Age Group
plt.subplot(4, 1, 2)
sns.barplot(x='AgeGroup', y='Survived', data=data, palette='PuBu',hue='AgeGroup')
plt.xlabel('Age Group')
plt.ylabel('Average Survival Rate')
plt.title('Survival Rate by Age Group')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
```

```python
# Bar plot for Survival vs Pclass
plt.subplot(4, 1, 3)
sns.barplot(x='Pclass', y='Survived', data=data, palette='PuBu',hue='Pclass' )
plt.xlabel('Passenger Class')
plt.ylabel('Average Survival Rate')
plt.title('Survival Rate by Passenger Class')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
```

```python
# Bar plot for Survival vs Embarked
plt.subplot(4, 1, 4)
sns.barplot(x='AgeGroup', y='Survived', data=data, palette='PuBu',hue='AgeGroup')
plt.xlabel('Embarked')
plt.ylabel('Average Survival Rate')
plt.title('Survival Rate by Embarked Port')
plt.xticks(ticks=[0, 1, 2], labels=['C', 'Q', 'S'])  # Map numeric codes back to labels

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.tight_layout()
plt.show()
```
